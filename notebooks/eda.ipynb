```python
import json
import pandas as pd
import matplotlib.pyplot as plt

with open("../data/sale_2025-12-18.json", encoding="utf-8") as f:
    data = json.load(f)
df = pd.json_normalize(data)
df.head()
```

## Visualization


```python
missing = df.isna().sum().sort_values(ascending=False)
missing = missing[missing > 0]

missing.plot(kind="bar")
plt.title("Missing values per column")
plt.ylabel("Count")
plt.show()
```

```python
df["price"].dropna().apply(lambda x: x if x > 0 else None).hist(bins=50)
plt.title("Price distribution")
plt.xlabel("Price")
plt.show()
```

```python
price_province = (
    df.groupby("address.province")["price_per_spm"]
      .median()
      .sort_values(ascending=False)
      .head(10)
)

price_province.plot(kind="barh")
plt.title("Top 10 provinces by median price per m²")
plt.xlabel("Price per m²")
plt.show()
```
